# Matplotlib

### Task 1

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'svg'

title_font = {
    'fontsize': 16,
    'fontweight': 'bold',
    'color': 'gray',
    'family': 'serif',
}
```

```python
x = np.arange(1, 8)
y = np.array([3.5, 3.8, 4.2, 4.5, 5, 5.5, 7])

plt.plot(x, y, label='f(x)')
plt.title('График функции', fontdict=title_font)
plt.xlabel('Переменная x', fontdict=title_font)
plt.ylabel('Функция f(x)', fontdict=title_font)
plt.legend(loc='lower right')
plt.show()
```

```python
plt.scatter(x, y, label='f(x)')
plt.title('Диаграмма разброса', fontdict=title_font)
plt.xlabel('Переменная x', fontdict=title_font)
plt.ylabel('Функция f(x)', fontdict=title_font)
plt.legend(loc='lower right')
plt.show()
```

### Task 2

```python
t = np.linspace(0, 10, 51)
f = np.cos(t)

plt.plot(t, f, label='График f(t)', color="green")
plt.title('График f(t)', fontdict=title_font)
plt.xlabel('Значения t', fontdict=title_font)
plt.ylabel('Значения f', fontdict=title_font)
plt.axis([0.5, 9.5, -2.5, 2.5])
plt.legend(loc='lower right')
plt.show()
```

### Task 3

```python
x = np.linspace(-3, 3, 51)
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)

fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()

fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)

ax1.plot(x, y1)
ax1.set_xlim([-5, 5])
ax1.set_title('График y1', fontdict=title_font)
ax2.plot(x, y2)
ax2.set_title('График y2', fontdict=title_font)
ax3.plot(x, y3)
ax3.set_title('График y3', fontdict=title_font)
ax4.plot(x, y4)
ax4.set_title('График y4', fontdict=title_font)
```

### Task 4

```python
plt.style.use('fivethirtyeight')
creditcard = pd.read_csv('creditcard.csv')
class_counts = creditcard['Class'].value_counts()
class_counts
```

```python
class_counts.plot(kind='bar')
plt.show()
```

```python
class_counts.plot(kind='bar', logy=True)
plt.show()
```

```python
class_0 = creditcard.loc[creditcard['Class'] == 0, ['V1']]
class_1 = creditcard.loc[creditcard['Class'] == 1, ['V1']]
plt.hist(class_0['V1'], bins=20, density=True, alpha=0.5, label='Class 0', color='grey')
plt.hist(class_1['V1'], bins=20, density=True, alpha=0.5, label='Class 1', color='red')
plt.title('Транзакции', fontdict=title_font)
plt.xlabel('V1', fontdict=title_font)
plt.ylabel('Плотность распределения', fontdict=title_font)
plt.legend(loc='upper left')
plt.show()
```
