# Обучение с учителем Scikit-learn

```python
import numpy as np
import pandas as pd

from sklearn.datasets import load_boston
```

### Task 1

```python
boston = load_boston()
data = boston["data"]
feature_names = boston["feature_names"]
target = boston["target"]

X = pd.DataFrame(data, columns=feature_names)
Y = pd.DataFrame(target, columns=["price"])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.30, random_state=42)
lr = LinearRegression()
lr.fit(X_train, Y_train)
```

```python
y_pred = lr.predict(X_test)
check_test = pd.DataFrame({
    "Y_test": Y_test["price"], 
    "Y_pred": y_pred.flatten()})

check_test
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(check_test["Y_pred"], check_test["Y_test"])
print(f'R2 равно:{r2}')
```

### Task 2

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators=1000, max_depth=12, random_state=42)
model.fit(X_train, Y_train.values[:, 0])
```

```python
y_pred = model.predict(X_test)
check_test = pd.DataFrame({
    "Y_test": Y_test["price"], 
    "Y_pred": y_pred.flatten()})

check_test
```

```python
r2_tree = r2_score(check_test["Y_pred"], check_test["Y_test"])
print(f'R2 равно:{r2_tree}')
```

```python
if r2 > r2_tree:
    print(f'Модель линейной регрессии работает лучше')
else:
    print(f'Модель RandomForestRegressor работает лучше')
```

### Task 4

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, train_test_split
```

```python
df = pd.read_csv('creditcard.csv')
df_counts = df['Class'].value_counts(normalize=True)
df_counts
```

```python
df.info()
```

```python
pd.options.display.max_columns=100
df.head(10)
```

```python
X = df.drop('Class', axis=1)
y = df['Class']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100, stratify=y)
print(f'X_train shape: {X_train.shape}')
print(f'X_test shape: {X_test.shape}')
print(f'y_train shape: {y_train.shape}')
print(f'y_test shape: {y_test.shape}')
```

```python
parameters = [{
    'n_estimators': [10, 15], 
    'max_features': np.arange(3, 5), 
    'max_depth': np.arange(4, 7)
}] 
clf = GridSearchCV(
    estimator=RandomForestClassifier(random_state=100),
    param_grid=parameters,
    scoring='roc_auc',
    cv=3,
)
clf.fit(X_train, y_train)
```

```python
clf.best_params_
```

```python
clf = RandomForestClassifier(max_depth=6, max_features=3, n_estimators=15)

clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import roc_auc_score

y_pred = clf.predict_proba(X_test)
y_pred_proba = y_pred[:, 1]
roc_auc_score(y_test, y_pred_proba)
```

### Task 5

```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'svg'

title_font = {
    'fontsize': 16,
    'fontweight': 'bold',
    'color': 'gray',
    'family': 'serif',
}
```

```python
digits = load_digits()
X = digits['data']
y = digits['target']
plt.figure(figsize=(10, 6))
for i in range(10):
    plt.subplot(2, 5, i + 1)
    plt.imshow(X[i,:].reshape([8,8]));
```

```python
parameters = [{
    'n_estimators': [10, 15], 
    'max_features': [1, 2, 3, 5, 10, 20 ,30, 50, 64], 
    'max_depth': [1, 2, 3, 5, 10, 20, 25, 30, 40, 50, 64]
}] 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
clf = GridSearchCV(
    estimator=RandomForestClassifier(random_state=100),
    param_grid=parameters,
    cv=3,
)
clf.fit(X_train, y_train)
```

```python
clf.best_params_
```

```python
clf = RandomForestClassifier(max_depth=10, max_features=10, n_estimators=15)

clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = clf.predict(X_test)
accuracy_score(y_test, y_pred)
```
